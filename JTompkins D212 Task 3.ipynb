```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
df = pd.read_csv('teleco_market_basket.csv',dtype={'locationid':np.int64})
df.head(5)
```

```python
df.iloc[3]
```

```python
df = df[df['Item01'].notna()]
df.reset_index(drop = True, inplace = True)
```

```python
df.shape
```

```python

df.head(5)
```

```python
df_list = []
for i in range(0,7501):
    df_list.append([str(df.values[i,j]) for j in range(0,20)])
```

```python
encoder = TransactionEncoder()
onehot = encoder.fit(df_list).transform(df_list)
final_df = pd.DataFrame(onehot, columns = encoder.columns_)
```

```python
frequent_itemsets = apriori(new_df, use_colnames = True, min_support = 0.02)
```

```python
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets, metric = 'lift', min_threshold = 1.0)
rules
```

```python
top_rules = rules.sort_values(by=['lift'], ascending = False)
top_rules.head(3)
```
